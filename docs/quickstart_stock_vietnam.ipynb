## II. Danh sách mã chứng khoán

Lấy danh sách các cổ phiếu giao dịch trên các sàn Việt Nam (HOSE, HNX, UPCOM).

**Nguồn dữ liệu:** VCI (Chứng khoán Vietcap) - có thể dùng TCBS thay thế

### Khởi tạo đối tượng

```python
from vnstock import Listing

# Khởi tạo đối tượng Listing với nguồn VCI
listing = Listing(source='VCI')
```

### Danh sách tất cả mã cổ phiếu (Đơn giản)

```python
# Lấy danh sách đơn giản tất cả mã chứng khoán
all_symbols = listing.all_symbols()
print(f"Tổng số mã: {len(all_symbols)}")
all_symbols.head(10)
```

### Danh sách cổ phiếu theo sàn giao dịch (Đầy đủ)

```python
# Lấy thông tin chi tiết cho tất cả mã theo sàn giao dịch
# Bao gồm cả mã đang và đã ngừng niêm yết
by_exchange = listing.symbols_by_exchange()
by_exchange.head()
```

### Danh sách cổ phiếu theo nhóm

```python
# Liệt kê tất cả mã chứng khoán theo nhóm phân loại. Ví dụ HOSE, VN30, VNMidCap, VNSmallCap, VNAllShare, VN100, ETF, HNX, HNX30, HNXCon, HNXFin, HNXLCap, HNXMSCap, HNXMan, UPCOM, FU_INDEX (mã chỉ số hợp đồng tương lai), CW (chứng quyền)
listing.symbols_by_group('VN30')
```

### Danh sách theo ngành (ICB)

```python
# Lấy danh sách cổ phiếu được nhóm theo ngành ICB
by_industry = listing.symbols_by_industries()
by_industry.head()
```

### Danh sách mã ngành theo mã icb

```python
listing.industries_icb()
```

### Danh sách mã hợp đồng tương lai

```python
listing.all_future_indices()
```

### Danh sách mã trái phiếu chính phủ

```python
listing.all_government_bonds()
```

### Danh sách bộ chỉ số

```python
listing.all_indices()
```

## Quote: Giá chứng khoán

```python
from vnstock import Quote
quote = Quote(symbol='ACB', source='VCI')
```

#### Giá lịch sử

##### Cổ phiếu

```python
# Khung thời gian ngày: 1D hay d cho interval
df = quote.history(start='2024-01-01', end='2025-11-10', interval='d')
df
```

```python
# Khung thời gian phút: 1m hoặc m cho interval
df = quote.history(start='2025-01-02', end='2025-11-10', interval='1m')
df
```

```python
# Khung thời gian giờ: 1H hoặc h cho interval
df = quote.history(start='2025-01-02', end='2025-11-10', interval='h')
df
```

#### Intraday: dữ liệu khớp lệnh

## Finance: Báo cáo tài chính

```python
from vnstock import Finance
finance = Finance(symbol='ACB', source='VCI')
```

### Bảng cân đối kế toán

```python
finance.balance_sheet(period='year', lang='vi', dropna=True)
```

```python
finance.balance_sheet(period='quarter', lang='en', dropna=True).head()
```

### Báo cáo kết quả kinh doanh

```python
finance.income_statement(period='year', lang='vi', dropna=True).head()
```

```python
finance.income_statement(period='quarter', lang='en', dropna=True).head()
```

### Báo cáo lưu chuyển tiền tệ

```python
finance.cash_flow(period='year', dropna=True).head()
```

```python
finance.income_statement(period='quarter', lang='en', dropna=True).head()
```

### Chỉ số tài chính

```python
finance.ratio(period='year', lang='vi', dropna=True).head()
```

```python
finance.ratio(period='quarter', lang='en', dropna=True).head()
```
